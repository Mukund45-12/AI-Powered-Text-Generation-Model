```python
import numpy as np # linear algebra
import pandas as pd
```

```python
data = pd.read_json('/content/drive/MyDrive/Prodigy/News_Category_Dataset_v3.json/News_Category_Dataset_v3.json', lines=True)

# Step 3: Check the dataset
print(data.shape)
data.head()
```

```python
data['category'].value_counts()
```

```python
desired_category=['COMEDY','SPORTS','CRIME','EDUCATION']
data_new=data[data['category'].isin(desired_category)]
```

```python
data_new.head()
```

```python
data_new.shape
```

```python
data_new['category'].value_counts()
```

```python
min_samples=1014
df_comedy=data_new[data_new.category=='COMEDY'].sample(min_samples,random_state=2020)
df_SPORTS =data_new[data_new.category=='SPORTS'].sample(min_samples,random_state=2020)
df_CRIME  =data_new[data_new.category=='CRIME'].sample(min_samples,random_state=2020)
df_EDUCATION =data_new[data_new.category=='EDUCATION'].sample(min_samples,random_state=2020)
```

```python
df_balanced=pd.concat([df_comedy,df_SPORTS,df_CRIME,df_EDUCATION],axis=0)
```

```python
df_balanced.category.value_counts()
```

```python
#encoder
df_balanced['category_enc']=df_balanced['category'].map({
    'COMEDY':0,
    'SPORTS':1,
    'CRIME':2,
    'EDUCATION':3

})
```

```python
df_balanced.head()
```

```python
df_balanced.drop('category',axis=1,inplace=True)
```

```python
df_balanced.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(
        df_balanced.headline,
        df_balanced.category_enc,
        test_size=0.2,
        random_state=2023,
        stratify=df_balanced.category_enc)
```

```python
x_train.shape
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report
clf=Pipeline([
    ('vectorize_bow',CountVectorizer(ngram_range=(1,1))),
    ('Multi Nb',MultinomialNB())
])
clf.fit(x_train,y_train)
y_pred=clf.predict(x_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")

def preprocess(text):
    doc = nlp(text)
    filtered_tokens = []
    for token in doc:
        if token.is_stop or token.is_punct:
            continue
        filtered_tokens.append(token.lemma_)
    return " ".join(filtered_tokens)



df_balanced['text_pre'] = df_balanced['headline'].apply(preprocess)
```

```python
df_balanced.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(
        df_balanced.text_pre,
        df_balanced.category_enc,
        test_size=0.2,
        random_state=2023,
        stratify=df_balanced.category_enc)
```

```python
# Define and train the pipeline
clf = Pipeline([
    ('vectorize_bow', CountVectorizer(ngram_range=(1, 2))),
    ('Multi Nb', MultinomialNB())
])
clf.fit(x_train, y_train)

# Predict on the test set
y_pred = clf.predict(x_test)
```

```python
print(classification_report(y_test,y_pred))
```
